##Rock Paper Scissors: 

We have to choose rock or paper or scissor then the computer will randomly choose from the inputs and finally it will compare 
both the answers and the stronger one wins



```python
#importing random library that is in-built module of Python which is used to generate random numbers.

import random

# using while loop to execute a block of code repeatedly until given boolean condition evaluated to False, so that if player do not pick from options it will continue forever


while True:
    choices =["rock","paper","scissors"]
    
#computer will choose randomly from the given options
    
    computer=random.choice(choices)
    player=None
    
#Asking player to choose from given options
#Also using lower() function to convert upper cases to lower if player types in upper case
    
    while player not in choices:
        player=input("rock,paper,or scissors? : ").lower()

#Now using if and else condition to compare the choices of player and computer
#Here if computer and player chooses same options then there will be a tie

        if player == computer:
            print("computer : ",computer)
            print("player : ",player)
            print("Tie!")
            
#If player chooses rock and computer chooses paper player loses but if computer chooses scissors then player wins
            
        elif player =="rock":
            if computer == "paper":
                print("computer : ", computer)
                print("player : ",player)
                print("You lose!")
            if computer == "scissors":
                print("computer : ", computer)
                print("player : ",player)
                print("You win!")
                
#If player chooses scissors and computer chooses rock player loses but if computer chooses paper player wins              

        elif player =="scissors":
            if computer == "rock":
                print("computer : ", computer)
                print("player : ",player)
                print("You lose!")
            if computer == "paper":
                print("computer : ", computer)
                print("player : ",player)
                print("You win!")
              
#If player chooses paper and computer chooses scissors player loses but if computer chooses rock player wins
                
        elif player =="paper":
            if computer == "scissors":
                print("computer : ", computer)
                print("player : ",player)
                print("You lose!")
            if computer == "rock":
                print("computer : ", computer)
                print("player : ",player)
                print("You win!")
                
#Creating a temporary variable that will ask player that will he play again or not
#Again using lower() function to convert uppercase into lowercase if player types in upper case        
    play_again=input("play again? (yes/no):").lower()
#now if player wants to play it will start again other wise if not chooses to play the game then it will end   
    if play_again !="yes":
#Using the break statement to terminate the loop if player does not wants to play
        break

#At last by the time of quitting game it will show bye
        
print("Bye!")


        
```
